```python
def filter_df_by_common_column(df1, df2, common_column="name"):
    # Get the unique values in the "name" column of the first DataFrame
    unique_names = set(df1[common_column])

    # Filter the second DataFrame to keep only rows with "name" values present in the first DataFrame
    filtered_df2 = df2[df2[common_column].isin(unique_names)]

    return filtered_df2
```
