# Masked wordcloud

```python
from os import path
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np
```

```python
text1 = open('C:/Users/Ajey Kumar/Desktop/TA/alice.txt').read()
```

```python
print(text1)
```

```python
# read the mask image
alice_mask = np.array(Image.open("C:/Users/Ajey Kumar/Desktop/TA/alice_mask.png"))
```

```python
stopwords = set(STOPWORDS)
stopwords.add("said")
```

```python
wc = WordCloud(background_color="white", max_words=2000, mask=alice_mask,
               stopwords=stopwords, contour_width=3, contour_color='steelblue')
```

```python
# generate word cloud
wc.generate(text)
```

```python
# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.show()
```

```python
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
